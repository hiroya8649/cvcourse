```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nadia = cv2.imread('DATA/Nadia_Murad.jpg', 0)
denis = cv2.imread('DATA/Denis_Mukwege.jpg', 0)
solvay = cv2.imread('DATA/solvay_conference.jpg', 0)
```

```python
plt.imshow(solvay, 'gray')
```

```python
face_cascade = cv2.CascadeClassifier('DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img)
    
    for (x, y, w, h) in face_rects:
        cv2.rectangle(face_img, (x,y), (x+w, y+h), (255,255,0), 10)
    return face_img
```

```python
res = detect_face(denis)
```

```python
plt.imshow(res, 'gray')
```

```python
eye_cascade = cv2.CascadeClassifier('DATA/haarcascades/haarcascade_eye.xml')
```

```python
def detect_eye(img):
    face_img = img.copy()
    face_rects = eye_cascade.detectMultiScale(face_img)
    
    for (x, y, w, h) in face_rects:
        cv2.rectangle(face_img, (x,y), (x+w, y+h), (255,255,0), 10)
    return face_img
```

```python
res = detect_eye(nadia)
```

```python
plt.imshow(res, 'gray')
```

```python
cap = cv2.VideoCapture('DATA/video_capture.mp4')
while True:
    ret, frame = cap.read(0)
    frame = detect_face(frame)
    cv2.imshow('Video', frame)
    k = cv2.waitKey(1)
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
