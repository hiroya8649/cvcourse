```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank = np.zeros((600, 600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank, text="ABCDE", org = (50, 300), fontFace=font, fontScale=5, color=(255,255,255),thickness=30)
    return blank
```

```python
def show_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,"gray")
```

```python
i = load_img()
show_img(i)
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
res = cv2.erode(i, kernel, iterations=5)
show_img(res)
```

```python
white_noise = np.random.randint(low=0, high=2, size=(600,600))
```

```python
white_noise = white_noise * 255
```

```python
noise_img = white_noise + i
show_img(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)
show_img(opening)
```

```python
black_noise = np.random.randint(low=0, high=2, size=(600,600))
```

```python
black_noise = black_noise*-255
```

```python
black_noise_img = black_noise+i
```

```python
black_noise_img[black_noise_img==-255]=0
```

```python
show_img(black_noise_img)
```

```python
res = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernel)
show_img(res)
```

```python
gradient = cv2.morphologyEx(i, cv2.MORPH_GRADIENT, kernel)
show_img(gradient)
```
