```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('DATA/sammy_face.jpg')
```

```python
edges = cv2.Canny(img, threshold1=126,threshold2=126)
```

```python
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
lower = int(max(0,0.7*med_val))
upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(img, threshold1=lower,threshold2=upper+50)
plt.imshow(edges)
```
