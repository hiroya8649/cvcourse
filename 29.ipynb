```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def show_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,"gray")
```

```python
img = cv2.imread("DATA/sudoku.jpg", 0)
show_img(img)
```

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)
```

```python
sobel = sobelx+sobely
show_img(sobel)
```

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
show_img(laplacian)
```

```python
ret, th1 = cv2.threshold(sobel, 220, 255, cv2.THRESH_BINARY)
k = np.ones((3,3))
th1 = cv2.morphologyEx(th1, cv2.MORPH_OPEN, k)
show_img(th1)
```

```python
kernel = np.ones((4,4),np.uint8)
gradient = cv2.morphologyEx(sobel, cv2.MORPH_GRADIENT, kernel)
show_img(gradient)
```
